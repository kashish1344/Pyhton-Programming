```python
import bs4
import urllib.request as url

path= 'https://www.flipkart.com/search?q=tv+smart+tv'
res= url.urlopen(path)
page = bs4.BeautifulSoup(res, 'lxml')
titles = page.find_all('div',class_='_4rR01T')          #query

priceList=page.find_all('div',class_='_30jeq3 _1_WHN1')

for i in range(len(titles)):
    print(titles[i].text)
    print(priceList[i].text)
    print('*' * 5)
```

```python
import bs4
import urllib.request as url

for i in range(4):
    print("Collection of Data of page: {}".format(i+1))
    path= 'https://www.flipkart.com/search?q=tv+smart+tv&page={}'.format(i+1)
    res= url.urlopen(path)
    page = bs4.BeautifulSoup(res, 'lxml')
    titles = page.find_all('div',class_='_4rR01T')          #query

    priceList=page.find_all('div',class_='_30jeq3 _1_WHN1')

    for j in range(len(titles)):
        print(titles[j].text)
        print(priceList[j].text)
        print('*' * 50)
```

```python
import bs4
import urllib.request as url

for i in range(4):
    print("Collecting data from page : {}".format(i+1))
    path = 'https://www.flipkart.com/search?q=tv+smart&page={}'.format(i+1)
    res = url.urlopen(path)
    page = bs4.BeautifulSoup(res, 'lxml')
    titles = page.find_all('div',class_='_4rR01T')

    priceList = page.find_all('div', class_='_30jeq3 _1_WHN1')
    rating= page.find_all('div', class_='_3LWZlK')
    offer= page.find_all('div', class_='_3Ay6Sb')
    for j in range(len(titles)):
        print("Mobile Name: ", titles[j].text)
        print("Price: ", priceList[j].text)
        print("Rating: ", rating[j].text)
        print("Offer Available: ", offer[j].text)
        print('*' * 50)
```

```python
import bs4
import urllib.request as url

job_name=input("Enter Your Profession: ")
job_name=job_name.replace(' ','-')
print("Job For ",job_name)
path='https://www.freshersworld.com/jobs/jobsearch/{}-jobs'.format(job_name)
res=url.urlopen(path)
page= bs4.BeautifulSoup(res, 'lxml')
job=page.find_all('div', class_='col-md-12 col-xs-12 col-lg-12 padding-none margin-top')
loc=page.find_all('div', class_='col-md-5 col-xs-5 col-lg-5 padding-none')

for i in range(len(job)):
    print(job[i].text)
    print()
    print('Location: ',loc[i].text)
    print('*' * 50)
```
